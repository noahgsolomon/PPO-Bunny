```python
import torch


class BobNet(torch.nn.Module):
    def __init__(self):
        super(BobNet, self).__init__()
        self.l1 = torch.nn.Linear(122, 64)  # Input layer with 122 features and 64 hidden units
        self.l2 = torch.nn.Linear(64, 4)    # Output layer with 64 hidden units and 4 outputs

    def forward(self, x):
        x = self.l1(x)
        x = torch.nn.ReLU()(x)
        x = self.l2(x)
        return x
```
